EDA Y ETL DE DATOS POBLACIONALES PARA RESPALDAR AN√ÅLISIS

```python
import pandas as pd
```

```python
# Ruta del archivo Excel
archivo_excel = 'datos_poblacionales_bancom.xlsx'

# Cargar solo la primera hoja en un DataFrame
df = pd.read_excel(archivo_excel, sheet_name=0) 
df
```

```python

df = df.rename(columns= {'Data Source':'Country Name', 'Indicadores del desarrollo mundial' : 'Country Code', 'Unnamed: 2':'2018', 'Unnamed: 3':'2019','Unnamed: 4' :'2020','Unnamed: 5' :'2021','Unnamed: 6':'2022','Unnamed: 7':'2023'})
df.head() 
```

```python
# Eliminar todas las columnas desde la columna 9 en adelante
df = df.drop(df.columns[8:], axis=1)
df
```

```python
# Eliminar las 3 primeras filas
df = df.drop(index=[0, 1, 2])
```

BOXPLOT

```python
# Seleccionar solo las columnas num√©ricas
columnas_numericas = df.select_dtypes(include='number')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("pastel")

# Crear el boxplot para las columnas num√©ricas
plt.figure(figsize=(10, 6))
sns.boxplot(data=columnas_numericas)

# T√≠tulo del gr√°fico
plt.title("Boxplot de columnas num√©ricas")

# Mostrar el gr√°fico
plt.show()
```
